# lkapi quickstart

The fastest way to pull your Lightkeeper portfolio data into Python. For full setup instructions (installing Python, launching Jupyter, etc.), see [`docs/getting-started.md`](../../docs/getting-started.md) in this repository.

**To use this notebook:**
1. Run the cells in order (click a cell, press `Shift+Enter`).
2. In the second code cell, replace `PASTE_YOUR_GRID_URL_HERE`, `YOUR_CLIENT_ID`, and `YOUR_CLIENT_SECRET` with your own values.

## 1. Install lkapi

Run this once. If `lkapi` is already installed, this is a quick no-op.

```python
pip install lkapi
```

## 2. Get your grid URL

In the Lightkeeper UI, open the grid (data view) you want data from, then go to **Grid > Api Routes** and copy the URL shown there. You'll also need the `client_id` and `client_secret` Lightkeeper gave you.

```python
import lkapi

frames = lkapi.get_grid_data(
    url="PASTE_YOUR_GRID_URL_HERE",
    username="YOUR_CLIENT_ID",
    password="YOUR_CLIENT_SECRET",
)

frames['rollup']
```

## 3. What you got back

`frames` is a dictionary of pandas DataFrames:

- `frames['rollup']`: one row per rollup (e.g. per issuer, per sector), summarized for the whole date range requested.
- `frames['time']`: one row per time period (e.g. per day), summarized across the portfolio.
- `frames['total']`: the overall totals.
- `frames['portfolio']`: details about the portfolio itself (available dates, last update time).
- `frames['request']`: details about the request made (useful for troubleshooting).

```python
frames['time']
```

> **Why is `frames['time']` empty above?** The saved example this notebook was captured with used a `viewby=rollup` URL, which intentionally skips the day-by-day breakdown to make the request much faster — so `frames['time']` comes back empty on purpose. If your own URL uses `viewby=time` (the default), you'll see one row per period here instead. See the "speed tip: `viewby`" section of the [getting started guide](../../docs/getting-started.md) for details.

```python
frames['total']
```

## 4. Save to CSV (optional)

Want this in Excel? Save any of the frames above as a CSV file:

```python
frames['rollup'].to_csv('rollup.csv', index=False)
```
